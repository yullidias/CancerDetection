```python
import pandas as pd
from resultado import *
from avaliacao import *
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
datasetDF = readDataset('id')
datasetDF.replace('M', 1, inplace=True)
datasetDF.replace('B', 0, inplace=True)
```

```python
datasetDF.head()
```

```python
ml_methods_params = [DecisionTreeClassifier(min_samples_split=0.0002,random_state=1),
                     DecisionTreeClassifier(min_samples_split=0.25,random_state=1),
                     DecisionTreeClassifier(min_samples_split=0.5,random_state=1)]

expDecisionTree = Experimento(Fold.gerar_k_folds(datasetDF,5,CLASSE), ml_methods_params)
expDecisionTree.saveFolds('tree')
```

```python
ml_methods_params = []
for i in range(-1,7,2):
    ml_methods_params += [SVC(kernel='linear', C=2**i)]
expSVMlinear = Experimento(Fold.gerar_k_folds(datasetDF,5,CLASSE), ml_methods_params)
```

```python
ml_methods_params = []
for i in range(-1,7,2):
    for j in range(-1,5,2):
        ml_methods_params += [SVC(kernel='rbf', C=2**i, gamma=2**j)]
expSVMRBF = Experimento(Fold.gerar_k_folds(datasetDF,5,CLASSE), ml_methods_params)
```

```python
ml_methods_params = []
for i in range(2,10,2):
    ml_methods_params += [KNeighborsClassifier(n_neighbors=2**i)]
expKNN = Experimento(Fold.gerar_k_folds(datasetDF,5,CLASSE), ml_methods_params)
```

```python
def resultados(nome_metodo, experimento):
    np.set_printoptions(precision=2)
    print(" === " + str(nome_metodo) + " === ")
    print("Numero de folds: " + str(len(experimento.folds)))
    print("Macro f1: " + str(experimento.macro_f1_avg))
    
    for count in range(len(experimento.resultados)):
        print(str(count) + "\tprecisao: " 
              + str(experimento.resultados[count].precisao) 
              + "\trevocacao: " + str(experimento.resultados[count].revocacao))
        print(experimento.resultados[count].mat_confusao)
```
